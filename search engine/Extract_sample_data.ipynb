```python
df = pd.read_sql_query("""SELECT * FROM zipfiles""", conn)
df.head()
```

```python
import zipfile
import io
count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])

    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')
```

```python
df['file_content'] = df['content'].apply(decode_method)

df.head()
```

```python
import csv
```

```python
def select_random_40_percent_from_dataframe(df):
    # Calculate the number of elements to select for 30%
    num_elements_to_select = int(0.4 * len(df))

    # Use sample() function to randomly select 40% of the DataFrame
    selected_data = df.sample(n=num_elements_to_select)

    return selected_data
```

```python
select_random_40_percent_from_dataframe(df).to_csv("/content/drive/MyDrive/sample_subtitle_data.csv",quoting=csv.QUOTE_ALL,escapechar="\\")
```
